```python
third_s = []
with open('output.txt', 'r') as f:
    lines = f.readlines()
    for line in lines:
        third_s += [int(line.split()[2])]
```

```python
third_s_s = set(third_s)
for t in third_s_s:
    print(t, ":", third_s.count(t))
```

```python
with open('output1.txt', 'w') as f1:
    with open('output.txt', 'r') as f:
        lines = f.readlines()
        for line in lines:
            line1 = line.split()[0] + ' ' + line.split()[1] + ' ' + str(third_s.index(int(line.split()[2])))
            f1.write(line1+'\n')
```
